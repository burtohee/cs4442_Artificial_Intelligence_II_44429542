## Q1 (a)


$$
f(w) = a^T (w \circ b)
$$
$w ∈ R^n $ is an n-dimensional column vector

$a, b ∈ R^n$ be two n-dimensional vectors,  

$a ◦ b ∈ R^n$ be the element-wise product of two vectors.




$$
(w \circ b) = 
\begin{bmatrix}
w_1 b_1 \\
\vdots \\
w_2 b_2\\
w_n b_n
\end{bmatrix}

$$

$$
f(w) = a^T (w \circ b) = \sum_{i=1}^{n} a_i (w_i  b_i)
$$


### To computer gradient of f(w);  ∇f(w) 

$$

\nabla f(x) =
\begin{bmatrix}
\frac{\partial f}{\partial w_1} \\
\vdots \\
\frac{\partial f}{\partial w_n}
\end{bmatrix}


$$

$$
\frac{\partial f}{\partial w_j} = \frac{\partial}{\partial w_j} \sum_{i=1}^{n} a_i (w_i  b_i) = a_j b_j
$$
which  we  focus  on j-th component

$$ 
\nabla f(w)  
= \frac{\partial f}{\partial w_j} = \nabla f(w) = 
\begin{pmatrix}
\frac{\partial f(w)}{\partial w_1} \\
\frac{\partial f(w)}{\partial w_2} \\
\vdots \\
\frac{\partial f(w)}{\partial w_n}
\end{pmatrix}
= 
\begin{pmatrix}
a_1  b_1 \\
a_2 b_2 \\
\vdots \\
a_n  b_n
\end{pmatrix}
= (a∘b) 
\\
= a \odot b (\text{which ⊙ is element-wise multiplication operation})


$$

```python
# import sys
# !{sys.executable} -m pip install --user numpy
# !{sys.executable} -m pip install --user matplotlib
import random
import numpy as np
import matplotlib.pyplot as plt


def convert_potential_vec_mat(vec):
    shapes = np.shape(vec)
    if(len(shapes) == 1):
        return np.array([vec]).T
    return np.array(vec)

def load_data(file):
    potential_vec = np.loadtxt(file)
    return convert_potential_vec_mat(potential_vec)

def combine_two_matrix_cols(x, y):
    return np.hstack((x,y))

# add column 1s to left most of X
def add_col_1s_left (x_data):
    return combine_two_matrix_cols(np.ones((x_data.shape[0], 1)), x_data)
    # return np.hstack((np.ones((x_data.shape[0], 1)), x_data))

# build linear regression X
def get_linear_regression_formula_X (x_data):
    return add_col_1s_left(x_data)

# build polynomial regression X
def get_polynomial_regression_formula_X (x_data, order):
    X = np.empty([len(x_data), 0])
    for n in range(order,0,-1):
        x_N_power = np.power(x_data, n)
        X = np.column_stack((X, x_N_power))
    return add_col_1s_left(X)

# Compute the weight vector using the formula: w = (X^T * X)^-1 * X^T * y
def linear_regression_formula(X, Y):
    return np.dot((np.linalg.inv(np.dot(X.T,X))), np.dot(X.T,Y))

def move_row_to_last(matrix, x):
    rows = matrix.shape[0]
    if x >= rows:
        return matrix  # If x is out of range, return the original matrix
    else:
        indices = [i for i in range(rows) if i != x]  # Exclude xth row
        indices.append(x)  # Append xth row at the end
        return matrix[indices, :]

# polynomial function follow size W
def get_polynomial_function(W):
    W = move_row_to_last(W, 0)
    return np.polynomial.polynomial.Polynomial(W[::-1].flatten())

def get_predict(X_test, beta):
    return np.dot(X_test, beta)

def get_predict_with_i(W,X):
    return np.dot(W.T, X)

def get_cols(X, start, stop):
    return X[:, start: stop +1]

# cal error
def calculate_error(W, X, Y):
    # average_error =  np.mean((X.dot(W) - Y)**2)
    ## print(average_error) 
    # return average_error
    m = len(Y)  # Number of training examples
    total_error = 0
    for i in range(m):
        xi = convert_potential_vec_mat(X[i])
        # computer hw_x
        hw_x = get_predict_with_i(W, xi)
        yi = convert_potential_vec_mat(Y[i])
        error = ((hw_x - yi) ** 2).flatten()
        total_error += error
    average_error = total_error / m
    return average_error
    # return np.round(average_error, round)[0]

def get_rounded(data, round):
    return np.round(data, round)

def simple_plot(scx,scy, scc, scl, px,pf, pc, pl, yrange):
    ax = plt.gca()
    ax.set_ylim(yrange)
    plt.scatter(scx,scy, color=scc, label=scl)
    plt.plot(px, pf, color=pc, label=pl)

def simple_plot_2(scale, px,pf, pc, pl, marker):
    if scale:
        plt.xscale(scale)  # Set x-axis scale to logarithmic
        
    if marker:
        plt.plot(px, pf, color=pc, label=pl, marker=marker)
    else:
        plt.plot(px, pf, color=pc, label=pl)

def show_plot(xl, yl,t, grid):
    plt.xlabel(xl)
    plt.ylabel(yl)
    plt.title(t)
    plt.legend()
    plt.grid(grid)
    plt.show()


def get_I_hat(W):
    result = np.identity(len(W))
    result[0][0] = 0
    return result

# Compute the weight vector using the formula: w = (X^T * X)^-1 * X^T * y
def linear_regression_formula_l2_regu(X, Y, I, lambda_s):
    return np.dot((np.linalg.inv( np.dot(X.T,X) +  lambda_s * I)), np.dot(X.T,Y))

def shuffle_and_split(data, k):
    d_temp = []
    for d in data:
        d_temp.append(d)
    random.shuffle(d_temp)  # Shuffle the data in place

    n = len(d_temp)
    group_size = n // k
    remainder = n % k
    groups = []
    start_index = 0
    for i in range(k):
        # Adjust group size if there's a remainder, stop until i = remainder
        group_end = start_index + group_size + (1 if i < remainder else 0) 
        groups.append(data[start_index:group_end])
        start_index = group_end
    return groups

def matrix_delete_cols(a , start, stop):
    return np.delete(a, np.s_[start:stop], axis=1)    

def split_x_y(data_set):
    shuffle_x = [] 
    shuffle_y = []        
    m,n = data_set.shape
    y =get_cols(data_set , n - 1, n-1)
    x = get_cols(data_set , 0, n-1-1)
    shuffle_x.append(x)
    shuffle_y.append(y)   
    return shuffle_x[0] , shuffle_y[0]

def slice_method(a, index):
    a = a[:index] + a[index+1:]
    return a

def all_data_from_arrays(a):
    result = []
    for x in a:
        for x_i in x:
            result.append(x_i)
    return result


def five_fold_cross_validation_on_lambda_with_x_order(x_data, y_data, K, prior_knowledge_max, o_n):
    x_y_data = combine_two_matrix_cols(x_data, y_data)
    J_group = []
    for l_index in range(len(prior_knowledge_max)):
        error_va_group_k = []
        group_x_y_data = shuffle_and_split(x_y_data, K)
        for k in range(K):
                # data processing
                validation_set = group_x_y_data[k]
                training_set = slice_method(group_x_y_data, k)
                validation_set_np = np.array(validation_set)
                training_set_all_data = all_data_from_arrays(training_set)
                training_set_all_data_np = np.array(training_set_all_data)

                training_set_np_x,  training_set_np_y= split_x_y(training_set_all_data_np)
                validation_set_np_x,  validation_set_np_y= split_x_y(validation_set_np)
                

                X_poly_reg_training_set_np_x = get_polynomial_regression_formula_X(training_set_np_x, o_n)
                W_poly_reg_training_set_np_x = linear_regression_formula(X_poly_reg_training_set_np_x, training_set_np_y)


                X_poly_reg_va = get_polynomial_regression_formula_X(validation_set_np_x, o_n)
                W_poly_reg_va = linear_regression_formula(X_poly_reg_va, validation_set_np_y)

                W_poly_reg_regu_cv = linear_regression_formula_l2_regu(X_poly_reg_training_set_np_x, training_set_np_y, get_I_hat(W_poly_reg_training_set_np_x), prior_knowledge_max[l_index])
                error_va = calculate_error(W_poly_reg_regu_cv, X_poly_reg_va, validation_set_np_y)
                error_va_group_k.append(error_va)
        J_l_index = np.mean(error_va_group_k)
        J_group.append(J_l_index)
    return J_group

# Load the training data
x_data = load_data('hw1xtr.dat')
y_data = load_data('hw1ytr.dat')

# Load the test data
x_data_te = load_data('hw1xte.dat')
y_data_te = load_data('hw1yte.dat')

X_linear_reg = get_linear_regression_formula_X(x_data)
W_linear_reg = linear_regression_formula(X_linear_reg, y_data)
# print(X_linear_reg)
# print(W_linear_reg)

X_linear_reg_te = get_linear_regression_formula_X(x_data_te)
W_linear_reg_te = linear_regression_formula(X_linear_reg_te, y_data_te)
# print(X_linear_reg)
# print(W_linear_reg)

order_n = 1
X_poly_reg = get_polynomial_regression_formula_X(x_data, order_n)
W_poly_reg = linear_regression_formula(X_poly_reg, y_data)
# print(X_poly_reg)
# print(W_poly_reg)

X_poly_reg_te = get_polynomial_regression_formula_X(x_data_te, order_n)
W_poly_reg_te = linear_regression_formula(X_linear_reg_te, y_data_te)
# print(X_linear_reg)
# print(W_linear_reg)

polynomial_function_linear_reg = get_polynomial_function(W_poly_reg)
# print(polynomial_function_linear_reg)
x_i = 0
xp = np.linspace(-0.5, 4.5, 100)
yrange = [-2.5, 12]


# predict_values = get_predict(X_linear_reg, W_linear_reg)
# print(predict_values)


simple_plot(get_cols(x_data, x_i, x_i), y_data, 'blue', 'Training data', xp, polynomial_function_linear_reg(xp), 'red', 'Linear regression line', yrange)
show_plot( 'X', 'Y', 'Linear Regression on Training Data', True)     

round_decimals = 4
average_error = calculate_error(W_linear_reg, X_linear_reg, y_data)
print("The average error on the training set using Eq. (1): ",get_rounded(average_error, round_decimals)[0])

```

```python
# import sys
# !{sys.executable} -m pip install --user numpy
# !{sys.executable} -m pip install --user matplotlib
import random
import numpy as np
import matplotlib.pyplot as plt


def convert_potential_vec_mat(vec):
    shapes = np.shape(vec)
    if(len(shapes) == 1):
        return np.array([vec]).T
    return np.array(vec)

def load_data(file):
    potential_vec = np.loadtxt(file)
    return convert_potential_vec_mat(potential_vec)

def combine_two_matrix_cols(x, y):
    return np.hstack((x,y))

# add column 1s to left most of X
def add_col_1s_left (x_data):
    return combine_two_matrix_cols(np.ones((x_data.shape[0], 1)), x_data)
    # return np.hstack((np.ones((x_data.shape[0], 1)), x_data))

# build linear regression X
def get_linear_regression_formula_X (x_data):
    return add_col_1s_left(x_data)

# build polynomial regression X
def get_polynomial_regression_formula_X (x_data, order):
    X = np.empty([len(x_data), 0])
    for n in range(order,0,-1):
        x_N_power = np.power(x_data, n)
        X = np.column_stack((X, x_N_power))
    return add_col_1s_left(X)

# Compute the weight vector using the formula: w = (X^T * X)^-1 * X^T * y
def linear_regression_formula(X, Y):
    return np.dot((np.linalg.inv(np.dot(X.T,X))), np.dot(X.T,Y))

def move_row_to_last(matrix, x):
    rows = matrix.shape[0]
    if x >= rows:
        return matrix  # If x is out of range, return the original matrix
    else:
        indices = [i for i in range(rows) if i != x]  # Exclude xth row
        indices.append(x)  # Append xth row at the end
        return matrix[indices, :]

# polynomial function follow size W
def get_polynomial_function(W):
    W = move_row_to_last(W, 0)
    return np.polynomial.polynomial.Polynomial(W[::-1].flatten())

def get_predict(X_test, beta):
    return np.dot(X_test, beta)

def get_predict_with_i(W,X):
    return np.dot(W.T, X)

def get_cols(X, start, stop):
    return X[:, start: stop +1]

# cal error
def calculate_error(W, X, Y):
    # average_error =  np.mean((X.dot(W) - Y)**2)
    ## print(average_error) 
    # return average_error
    m = len(Y)  # Number of training examples
    total_error = 0
    for i in range(m):
        xi = convert_potential_vec_mat(X[i])
        # computer hw_x
        hw_x = get_predict_with_i(W, xi)
        yi = convert_potential_vec_mat(Y[i])
        error = ((hw_x - yi) ** 2).flatten()
        total_error += error
    average_error = total_error / m
    return average_error
    # return np.round(average_error, round)[0]

def get_rounded(data, round):
    return np.round(data, round)

def simple_plot(scx,scy, scc, scl, px,pf, pc, pl, yrange):
    ax = plt.gca()
    ax.set_ylim(yrange)
    plt.scatter(scx,scy, color=scc, label=scl)
    plt.plot(px, pf, color=pc, label=pl)

def simple_plot_2(scale, px,pf, pc, pl, marker):
    if scale:
        plt.xscale(scale)  # Set x-axis scale to logarithmic
        
    if marker:
        plt.plot(px, pf, color=pc, label=pl, marker=marker)
    else:
        plt.plot(px, pf, color=pc, label=pl)

def show_plot(xl, yl,t, grid):
    plt.xlabel(xl)
    plt.ylabel(yl)
    plt.title(t)
    plt.legend()
    plt.grid(grid)
    plt.show()


def get_I_hat(W):
    result = np.identity(len(W))
    result[0][0] = 0
    return result

# Compute the weight vector using the formula: w = (X^T * X)^-1 * X^T * y
def linear_regression_formula_l2_regu(X, Y, I, lambda_s):
    return np.dot((np.linalg.inv( np.dot(X.T,X) +  lambda_s * I)), np.dot(X.T,Y))

def shuffle_and_split(data, k):
    d_temp = []
    for d in data:
        d_temp.append(d)
    random.shuffle(d_temp)  # Shuffle the data in place

    n = len(d_temp)
    group_size = n // k
    remainder = n % k
    groups = []
    start_index = 0
    for i in range(k):
        # Adjust group size if there's a remainder, stop until i = remainder
        group_end = start_index + group_size + (1 if i < remainder else 0) 
        groups.append(data[start_index:group_end])
        start_index = group_end
    return groups

def matrix_delete_cols(a , start, stop):
    return np.delete(a, np.s_[start:stop], axis=1)    

def split_x_y(data_set):
    shuffle_x = [] 
    shuffle_y = []        
    m,n = data_set.shape
    y =get_cols(data_set , n - 1, n-1)
    x = get_cols(data_set , 0, n-1-1)
    shuffle_x.append(x)
    shuffle_y.append(y)   
    return shuffle_x[0] , shuffle_y[0]

def slice_method(a, index):
    a = a[:index] + a[index+1:]
    return a

def all_data_from_arrays(a):
    result = []
    for x in a:
        for x_i in x:
            result.append(x_i)
    return result


def five_fold_cross_validation_on_lambda_with_x_order(x_data, y_data, K, prior_knowledge_max, o_n):
    x_y_data = combine_two_matrix_cols(x_data, y_data)
    J_group = []
    for l_index in range(len(prior_knowledge_max)):
        error_va_group_k = []
        group_x_y_data = shuffle_and_split(x_y_data, K)
        for k in range(K):
                # data processing
                validation_set = group_x_y_data[k]
                training_set = slice_method(group_x_y_data, k)
                validation_set_np = np.array(validation_set)
                training_set_all_data = all_data_from_arrays(training_set)
                training_set_all_data_np = np.array(training_set_all_data)

                training_set_np_x,  training_set_np_y= split_x_y(training_set_all_data_np)
                validation_set_np_x,  validation_set_np_y= split_x_y(validation_set_np)
                

                X_poly_reg_training_set_np_x = get_polynomial_regression_formula_X(training_set_np_x, o_n)
                W_poly_reg_training_set_np_x = linear_regression_formula(X_poly_reg_training_set_np_x, training_set_np_y)


                X_poly_reg_va = get_polynomial_regression_formula_X(validation_set_np_x, o_n)
                W_poly_reg_va = linear_regression_formula(X_poly_reg_va, validation_set_np_y)

                W_poly_reg_regu_cv = linear_regression_formula_l2_regu(X_poly_reg_training_set_np_x, training_set_np_y, get_I_hat(W_poly_reg_training_set_np_x), prior_knowledge_max[l_index])
                error_va = calculate_error(W_poly_reg_regu_cv, X_poly_reg_va, validation_set_np_y)
                error_va_group_k.append(error_va)
        J_l_index = np.mean(error_va_group_k)
        J_group.append(J_l_index)
    return J_group



# Load the training data
x_data = load_data('hw1xtr.dat')
y_data = load_data('hw1ytr.dat')

# Load the test data
x_data_te = load_data('hw1xte.dat')
y_data_te = load_data('hw1yte.dat')

# X_linear_reg = get_linear_regression_formula_X(x_data)
# W_linear_reg = linear_regression_formula(X_linear_reg, y_data)

# X_linear_reg_te = get_linear_regression_formula_X(x_data_te)
# W_linear_reg_te = linear_regression_formula(X_linear_reg_te, y_data_te)

# order_n = 4
# X_poly_reg = get_polynomial_regression_formula_X(x_data, order_n)
# W_poly_reg = linear_regression_formula(X_poly_reg, y_data)

# X_poly_reg_te = get_polynomial_regression_formula_X(x_data_te, order_n)
# W_poly_reg_te = linear_regression_formula(X_linear_reg_te, y_data_te)

# polynomial_function_linear_reg = get_polynomial_function(W_poly_reg)
# # print(polynomial_function_linear_reg)
x_i = 0
xp = np.linspace(-0.5, 4.5, 100)
yrange = [-2.5, 12]

# predict_values = get_predict(X_linear_reg, W_linear_reg)
# print(predict_values)


lambda_s = [0.01, 0.1, 1, 10, 100, 1000]
k_fold_n = 5
order_n = 4
J_group = five_fold_cross_validation_on_lambda_with_x_order(x_data, y_data, k_fold_n, lambda_s, order_n)

simple_plot_2('log', (lambda_s), J_group, 'blue', 'validation average error', marker='o')
show_plot("\u03BB", 'Error', 'five-fold cross-validation on \u03BB with 4th order polynomial regression', True)

lambda_error_min = {"min": float('inf')}
lambda_error_min_index = float('inf')
for j in range(len(J_group)):
    min = lambda_error_min["min"]
    if(J_group[j] < min):
        lambda_error_min["min"] = J_group[j]
        lambda_error_min_index = j
lambda_error_min = lambda_error_min["min"]
best_lambda_after_cv = lambda_s[lambda_error_min_index]

print("The best \u03BB after cross validation is: " + str(best_lambda_after_cv) + " with validation average error: " + str(lambda_error_min))
print("The best \u03BB after cross validation is the same as in (a)")
print("same \u03BB for training data in (a), not same \u03BB for testing data in (a)")

X_poly_reg_te = get_polynomial_regression_formula_X(x_data_te, order_n)
# W_poly_reg_te = linear_regression_formula(X_poly_reg_te, y_data_te)


X_poly_reg_tr_all = get_polynomial_regression_formula_X(x_data, order_n)
W_poly_reg_tr_add = linear_regression_formula(X_poly_reg_tr_all, y_data)
W_poly_reg_regu_after_cross_validation = linear_regression_formula_l2_regu(X_poly_reg_tr_all, y_data, get_I_hat(W_poly_reg_tr_add), best_lambda_after_cv)
polynomial_function_linear_reg = get_polynomial_function(W_poly_reg_regu_after_cross_validation)

plt.scatter(x_data_te,y_data_te, color="red", label="testing data")
simple_plot_2('linear', xp, polynomial_function_linear_reg(xp), 'blue', 'validation average error', marker=None)
show_plot('\u03BB', 'Y', 'l2 regulation-4th order polynomial on testing data', True) 

round_decimals = 4
average_error = calculate_error(W_poly_reg_regu_after_cross_validation, X_poly_reg_te, y_data_te)
print("The average error on the training set using Eq. (1): ", get_rounded(average_error, round_decimals)[0])
```
