$$
k(x, z) = a_1 k_1(x, z) - a_2 k_2(x, z)
\\
K_{ij} = \phi(x_i)^T\phi(x_j) = \phi(x_j)^T\phi(x_i) = K_{ji}; k_1(x, z), k_2(x, z) \text{ are valid kernel} 
\\

\alpha^T K \alpha = \sum_{i=1}^{m} \sum_{j=1}^{m} \alpha_i K_{ij} \alpha_j

\\
= \sum_{i=1}^{m} \sum_{j=1}^{m} \alpha_i \left( (a_1) \sum_{k=1}^{n`} \phi_{1k}(x_i) \cdot \phi_{1k}(x_j) \right) - \left( (a_2) \sum_{l=1}^{n``} \phi_{2l}(x_i) \cdot \phi_{2l}(x_j) \right) \alpha_j

## Q2 (c)


kernel $k_L$ for comparing the instances: and two strings u and v, 
you will count the number of subsequences up to length L that the two strings have in common
count the number of subsequences up to length L that the two strings
have in common. For example, if u = “car” and v = “ark”, and L = 1, then $k_L(u, v)$ = 2 (since ‘a’ and ‘r’ are common to the two strings). 
For the same strings, if L = 2, then kL(u, v) = 3, since ‘a’, ‘r’ and “ar” are in common. 
Let L = 2, is $k_L$ a valid kernel function?

when there are two strings in data, such as u, v

$x = [u,v], k_{L=2}$

then the kernel matrix K would be:


$$

k_{L=2}=
\begin{pmatrix}
    a & b \\
    b & e
\end{pmatrix}

$$

$$
b \leq min(a,e); a \geq b , e \geq b, ae \geq b^2; a > 0, e > 0, b > 0, ae \geq b^2, ae - b^2 \geq 0
$$

All upper lefe determinants are $\geq 0$

--

For 
$x = [u_1,u_2,...,u_{n}], k_{L=2}$, 
$\mathbf{c} = [c_1, c_2, \ldots, c_n]^T$
then the kernel matrix K would be:


$$

\mathbf{c}^T K \mathbf{c} = \sum_{i,j=1}^{n} c_i c_j k_L(u^{(i)}, u^{(j)})

\text{, for any pair of strings kL(u,v) is non-negative.}

then,

\mathbf{c}^T K \mathbf{c}  \geq 0
$$


then $k_{L=2}$ kernel matrix is positive semidefinite, then $k_{L=2}$ is valid kernel

```python
import numpy as np
import matplotlib.pyplot as plt

def convert_potential_vec_mat(vec):
    if(vec.ndim == 1):
        return np.array([vec])
    return np.array(vec)

def load_data(file):
    potential_vec = np.loadtxt(file)
    return convert_potential_vec_mat(potential_vec)

def reshape_imgs_matrix(data, matrix):
  x, y = matrix 
  if(data.ndim == 1):
     return data.reshape(x,y).T  # Reshape each image to 64x64
  return data.reshape(-1, x,y).T  # Reshape each image to 64x64

def show_image(img_data, cmap):
  if(cmap != None): 
    plt.imshow(img_data, cmap=cmap)
  else:
    plt.imshow(img_data)
  plt.show()

def data_mean(data, axis):
  # 0: cols 1: rows
  return np.mean(data, axis=axis)

def data_mean_center(data):
  mean = data_mean(data , 0)
  return data - mean

def data_std(data, axis):
  return np.std(data, axis=axis)

def data_std_center(data):
  std = data_std(data, 0)
  mean = data_mean(data, 0)
  return (data - mean) / std

def simple_plot(px,pf, pc, pl, marker):
    # if scale:
    #     plt.yscale(scale)  # Set x-axis scale to logarithmic
        
    if marker:
        plt.plot(px, pf, color=pc, label=pl, marker=marker)
    else:
        plt.plot(px, pf, color=pc, label=pl)

def simple_bar_plot(px,pf, pc, pl, width):
  plt.bar(px, pf, color =pc, width = width, label=pl)


def show_plot(xl, yl,t, grid):
    plt.xlabel(xl)
    plt.ylabel(yl)
    plt.title(t)
    plt.legend()
    plt.grid(grid)
    plt.show()

class pca_cus_class:

  e_values: np.ndarray
  e_vectors: np.ndarray

  def __init__(self):
    pass
    # if not isinstance(array, (list, array)):
    #     raise TypeError("Input must be a Python list or built-in array")
    # self.array = array
  
  def pca(self, data_centered):
    # data_centered = data_mean_center(data)

    data_cov = np.cov(data_centered.T)

    e_values, e_vectors = np.linalg.eigh(data_cov)

    # Sort eigenvalues and their eigenvectors in descending order
    e_ind_order = np.flip(e_values.argsort())
    self.e_values = e_values[e_ind_order]
    # re-order the columns, not rows
    self.e_vectors = e_vectors[:, e_ind_order]
    return self.e_values, self.e_vectors

  def max_eigenvalues(self):
     return np.max(self.e_values)
  
  @staticmethod
  def format_max_eigenvalues(largest_eigenvalue):
    # Format the largest eigenvalue in scientific notation with four decimal places
    return "{:.4e}".format(largest_eigenvalue)

  def pca_variance_percentage(self):
    # Compute the total variance
    total_variance = np.sum(self.e_values)
    each_variance = (self.e_values / total_variance )
    return each_variance
  
  def pca_variance_percentages(self):
    # Compute the total variance
    total_variance = np.sum(self.e_values)
    cumulative_variance = np.cumsum(self.e_values) / total_variance
    return cumulative_variance

  def determine_dimensionality(self, variance_threshold=0.95):
      cumulative_variance = self.pca_variance_percentages()
      n_components = np.argmax(cumulative_variance >= variance_threshold) + 1
      return n_components

  def pca_n_n_components(self, n_components, data):
    # Select the top 10 eigenvalues
    top_n_eigenvalues = self.e_values[:n_components]
    # Select the top 10 eigenvectors
    top_n_eigenvectors = self.e_vectors[:, :n_components]
    # Construct the image using the top 10 eigenvectors
    reconstructed_image = np.dot( np.dot(top_n_eigenvectors, top_n_eigenvectors.T), data)
    # reconstructed_image = np.matmul(data, top_n_eigenvectors.T)
    return reconstructed_image

def get_cols(X, start, stop):
    return X[:, start: stop]


def show_top_k_eigenvector(looks, figsize, e_vectors, matrix, k):
  # Create 5 subplots for each eigenvector
  fig, axs = plt.subplots(looks[0], looks[1], figsize=figsize)

  # Plot each eigenvector
  for i in range(k):
      e_v = e_vectors[:, i]
      axs[i].imshow(reshape_imgs_matrix(e_v, matrix), cmap='gray')
      axs[i].set_title(f"Eigenvector {i+1}")

  # Adjust layout
  plt.tight_layout()
  plt.show()


face_data = load_data("faces.dat")
face_data_mean_center = data_mean_center(face_data)

face_data_mean_center_m, face_data_mean_center_n = face_data.shape
select_data_m = int(face_data_mean_center_m * 0.4)

image_id = 200
image_index = 200 -1
image_200 = face_data[image_index]
image_matrix_size = [64,64]
image_200_reshape = reshape_imgs_matrix(image_200 , image_matrix_size)
cmap = "gray"
show_image(image_200_reshape, cmap)

```
